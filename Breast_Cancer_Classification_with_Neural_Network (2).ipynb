**Separating the Features and Target**

We now separate all the columns except the last one and dump those values into X, and the last column's values we dump into Y

```python
X = data_frame.drop(columns = 'label', axis =1)
Y = data_frame['label']
```

```python
print(X)
```

```python
print(Y)
```

**Separating Data into Training and Testing Data**

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size = 0.2,random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

**Standardise the data**

We fit the training data to X_train, and transforming it and storing it in X_train_std. Based on this scaling, we transform out test data.

The important thing to know here is that we must not fit the data to the test data. We must fit it to the training data.

Basically we are fitting and transforming the **Training Data** and then we use the same scaler to transform the test data.


```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)
```

"test_size = 0.2" tells us that 20% of the data is used as testing data.

**Building a Neural Netwrork**

We know that a NN contains several layers. The first layer is our input layer wherein the input is given, and the final layer is our output layer wherein the model predicts the output. We can have a single or multiple hidden layers.

To create the NN, we use two libraries nameply Tensorflow and Keras.

```python
print(X_train_std)
```

```python
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

Setting random seed will allow the model's accuracy to persist (stay the same) every single time we run the NN.

```python
#Setting up the layers of NN
model = keras.Sequential([
                          keras.layers.Flatten(input_shape=(30,)),
                          keras.layers.Dense(20, activation='relu'),
                          keras.layers.Dense(2, activation='sigmoid')
])
```

Here we have set up the 3 layers.

The hidden layer consists of 20 neurons, and the activation function is ReLU (Stads for Rectifies Linear Units).

The output layer consists of 2 neurons and the activation function is Sigmoid, which is just like the logistic regression function. Here the no. of neurons must be specific to the no of target values, in this caae we have only 2 targe/ouptut value, 0 or 1.

Dense indicates that neurons in that particular layer are connected to all the layers in the previous layer.

Since the input (The Data Frame) is in a 569 X 30 matrix form, we need to Flatten this, which means we need to convert in into a single Dimensional array which is very important.

```python
#Compiling tbe NN
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics =['accuracy']),
```

Loss function is basically the difference between true value and predicted value.

Our goal while compiling a NN is to decrease the loss as much as possible and increase the accuracy as much as possible.

```python
#Training the NN
history = model.fit(X_train_std, Y_train, validation_split = 0.1, epochs=10)
```

Here we use the standardised X_train_std instead of just X_train. This increases the accuracy.

**Visualising Accuracy and Loss**

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epochs')
plt.legend(['training data','val'], loc = 'lower right')
```

First epoch is labeled as 0...& so on

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.ylim(0.5, 1.0)
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epochs')
plt.legend(['training data','val'], loc = 'lower right')
```

The blue line tells us how the training data accuracy changes.
The Orange line tells us how validation accuracy changes.

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'upper right')
```

**Accuracy of the model on test data**

```python
loss, accuracy = model.evaluate(X_test_std, Y_test)
print(accuracy)
```

We feed the X_test_std and the NN wil predict the label or target for this standardised test data, and it will compare it with Y_test, which is our true data.

It gives the loss value and accuracy value.
